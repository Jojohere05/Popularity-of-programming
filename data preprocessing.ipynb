```python
df_copy=df.copy()
```

```python
irrelevant_col = ['Q120', 'Age', 'Employment', 'RemoteWork', 'LearnCodeOnline', 'YearsCodePro', 'OrgSize', 'PurchaseInfluence',
                  'TechList', 'BuyNewTool', 'Currency', 'CompTotal', 'OpSysPersonal use', 'OfficeStackAsyncHaveWorkedWith',
                  'OfficeStackAsyncWantToWorkWith', 'OfficeStackSyncHaveWorkedWith', 'OfficeStackSyncWantToWorkWith', 'NEWSOSites',
                  'SOVisitFreq', 'SOAccount', 'SOPartFreq', 'SOComm', 'SOAI', 'AISent', 'AIBen', 'AIToolCurrently Using',
                  'AIToolNot interested in Using', 'AINextVery different', 'AINextNeither different nor similar', 'AINextSomewhat similar',
                  'AINextVery similar', 'AINextSomewhat different', 'TBranch', 'ICorPM', 'Knowledge_1', 'Knowledge_2', 'Knowledge_3',
                  'Knowledge_4', 'Knowledge_5', 'Knowledge_6', 'Knowledge_7', 'Knowledge_8', 'Frequency_1', 'Frequency_2',
                  'Frequency_3', 'TimeSearching', 'TimeAnswering', 'Industry', 'SurveyLength', 'SurveyEase', 'ConvertedCompYearly']
df_copy = df_copy.drop(columns=irrelevant_col, axis=1)
missing_value1 = df_copy.isnull().sum().to_dict()
mis1 = {col: ((val / df_copy.shape[0]) * 100) for col, val in missing_value1.items()}
mis1 = {col: f"{round(val, 2)}%" for col, val in mis1.items()}
threshold = 0.50
missing_fraction = {col: float(val.strip('%')) / 100 for col, val in mis1.items()}
columns_to_drop = [col for col, val in missing_fraction.items() if val > threshold]


df_cleaned = df_copy.drop(columns=columns_to_drop, axis=1)

lower_threshold = 0.40
upper_threshold = 0.50

columns_to_drop_40_50 = [col for col, val in missing_fraction.items() if lower_threshold <= val < upper_threshold]

df_cleaned = df_cleaned.drop(columns=columns_to_drop_40_50, axis=1)

print(df_cleaned.info())
```

```python

total_rows_cleaned = df_cleaned.shape[0]

missing_values_cleaned = df_cleaned.isnull().sum().to_dict()

missing_percentage_cleaned = {col: (val / total_rows_cleaned) * 100 for col, val in missing_values_cleaned.items()}

missing_percentage_cleaned = {col: f"{round(val, 2)}%" for col, val in missing_percentage_cleaned.items()}

print(missing_percentage_cleaned)
```
